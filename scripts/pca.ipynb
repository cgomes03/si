```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
from si.io.csv_file import read_csv
from si.decomposition.pca import PCA
```

```python
#Load dataset

dataset = read_csv('../datasets/iris/iris.csv', features=True, label=True)
print("Original Shape:", dataset.X.shape)
print("Labels:", np.unique(dataset.y))
```

```python
#Execute PCA

pca = PCA(n_components=2)
pca.fit(dataset)

print("Variance Explained:", pca.explained_variance)
```

```python
# Transform data to and plot 

# Reducing dimensionality
dataset_reduced = pca.transform(dataset)

# Plotting
unique_labels = np.unique(dataset.y)
plt.figure(figsize=(8, 6))

for label in unique_labels:
    mask = (dataset.y == label)
    plt.scatter(dataset_reduced.X[mask, 0], dataset_reduced.X[mask, 1], label=label)

plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.legend()
plt.show()
```
