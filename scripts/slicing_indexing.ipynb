```python
import numpy as np
from si.io.csv_file import read_csv
```

```python
dataset = read_csv('../datasets/iris/iris.csv', features=True, label=True)
```

```python
# 1.2 - Select the penultimate variable (second-to-last column)
penultimate = dataset.X[:, -2]
print("Penultimate feature shape:", penultimate.shape)
print(penultimate)
```

```python
# 1.3 - Select the last 10 samples (rows)

last_10 = dataset.X[-10:, :]
# Compute the mean of each feature in these samples
means = last_10.mean(axis=0)
print("Means of last 10 samples:")
print(means)

# Print means feature by feature
for name, mean in zip(dataset.features, means):
    print(f"{name}: {mean:.3f}")
```

```python
# 1.4 - Find rows where all feature values are ≤ 6

mask = (dataset.X <= 6).all(axis=1)
filtered = dataset.X[mask]
print("Number of rows where all values ≤ 6:", filtered.shape[0])
```

```python
# 1.5 - Select rows where the label is not 'Iris-setosa'

mask = dataset.y != 'Iris-setosa'
filtered_X = dataset.X[mask]
filtered_y = dataset.y[mask]
print("Number of non-'Iris-setosa' rows:", filtered_X.shape[0])
print("Remaining classes:", set(filtered_y))
```
